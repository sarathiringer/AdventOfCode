```python
with open("input.txt", "r") as f:
    lines = f.readlines()
    lines = [l.strip() for l in lines]
```

```python
dirs = dict()
for i, line in enumerate(lines):
    if line == '$ cd ..':
        continue
    else:
        if line.startswith('$ cd'):
            dirs[line.split(' ')[2]] = list()
            for inner_line in lines[i+2:]:
                if inner_line.startswith('$ cd'):
                    break
                else:
                    dirs[line.split(' ')[2]].append(inner_line)
```

```python
for k, v in dirs.items():
    for i, item in enumerate(v):
        if item.startswith('dir'):
            d = item.split(' ')[1]
            dirs[k].extend(dirs[d])
```

```python
sums = dict()
for k, v in dirs.items():
    sums[k] = 0
    for file in v:
        if file.startswith('dir'):
            continue
        else: 
            sums[k] += int(file.split(' ')[0])
```

```python
below_10000 = 0
for k, v in sums.items():
    if v < 100000:
        below_10000 += v
```

```python
with open("input.txt", "r") as f:
    lines = f.readlines()
    lines = [l.strip() for l in lines]

dirs = dict()

def get_dir_size(lines):
    for i, line in enumerate(lines):
        if line.startswith('$ cd') and not line == '$ cd ..':
            directory = line.split(' ')[2]
        if line == '$ cd ..':
            start = lines.index(f'$ cd {directory}')
            end = i
            dir_content = lines[start:end]
            dir_content = dir_content[2:]
            content_size = 0
            for item in dir_content:
                if item[0].isdigit():
                    content_size += int(item.split(' ')[0])
            dirs[directory] = content_size
            lines[lines.index(f'dir {directory}')] = str(content_size)
            del lines[start:end+1]
            get_dir_size(lines)
```
